```python
import pandas as pd
```

```python
df_main = pd.read_excel("Salmonella_typhi_random_selection_July2018-Feb2023_biobank.xlsx",header=0)
df_main.head(2)
```

```python
df_sub = pd.read_excel("Salmonella_typhi_Extraction_Nanodrop_result_V1.xlsx",header=0)
df_sub.head(2)
```

```python
common_ids = pd.merge(df_main, df_sub, on='Specimen ID', how='inner')['Specimen ID']
```

```python
# Create a new column 'Matching' in the main DataFrame and initialize it as 'not match'
df_main['Matching'] = 'not match'
df_main.head(2)
```

```python

# Update 'Matching' column to 'match' where the 'Specimen ID' is present in the common_ids
df_main.loc[df_main['Specimen ID'].isin(common_ids), 'Matching'] = 'match'
df_main.head(2)
```

```python
# Save the updated DataFrame back to the same Excel file with the new column
df_main.to_excel("ok.xlsx", index=False)
```
