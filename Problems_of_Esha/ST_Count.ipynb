```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
library(readxl)
```

```r
# Specify the path to your excel file
excel_file <- "SA.xlsx"
```

```r
# Read the excel file into a data frame
data <- read_excel(excel_file)
head(data)
```

```r
# Count the occurrences of each ST
ST_counts <- data %>% count(ST)
head(ST_counts)
```

```r

# Create a stacked bar plot
ggplot(ST_counts, aes(x = "", y = n, fill = as.factor(ST))) +
  geom_bar(stat = "identity") +
  coord_polar("y") +
  labs(title = "ST Distribution in South Africa", x = "ST", y = "n") +
  scale_fill_discrete(name = "ST")
```
